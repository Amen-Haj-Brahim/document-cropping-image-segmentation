```python
import tensorflow as tf
import cv2 as cv
from pixellib.instance import instance_segmentation 
import matplotlib.pyplot as plt
```

```python
model=instance_segmentation()
model.load_model("mask_rcnn_coco.h5")
```

```python
imgs=[]
for i in range(1,8):
  imgs.append(cv.imread("test_images/"+str(i)+".jpg",cv.IMREAD_COLOR))
```

```python
plt.figure(figsize=(10,10))
for i in range(1,8):
  plt.subplot(2,4,i)
  plt.imshow(imgs[i-1])
  plt.axis("off")
  plt.title(str(i))
```

```python
res=model.segmentFrame(imgs[0],show_bboxes=True)
```
